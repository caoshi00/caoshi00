```python
##爬虫获取杭州市大型互联网公司名及信息地址
import requests
from bs4 import BeautifulSoup
target = 'https://www.lagou.com/gongsi/6-0-0-5,6'
pagenum=20
#获取target当中的所有页的内容
def get_page(url, num):
    page_list = []
    for i in range(1, num+1):
        formdata = {'type':'index', 'pn':i}
        try:
            req = requests.post(url, data = formdata)
            req.raise_for_status()
            req.encoding = req.apparent_encoding
            html = req.text
            bf = BeautifulSoup(html)
            texts = bf.find_all('div', id='company_list') #返回的是一个list
            if texts:
                page_list.append(str(texts[0]))
        except:
            print ('链接失败')
    return page_list
#获取页面中的公司链接dict
def get_url(html):
    li_bf = BeautifulSoup(html)
    li_texts = li_bf.find_all('li', class_='company-item')
    #li_texts[1]
    a_texts = []
    for li in li_texts:
        h_bf = BeautifulSoup(str(li))
        h3 = h_bf.find_all('h3', class_='company-name wordCut')
        a_bf = BeautifulSoup(str(h3[0]))
        a = a_bf.find_all('a')
        a_texts.append(a[0])
    href = dict()
    for a in a_texts:
        key = str(a.string)
        value = a.get('href')
        href[key] = value
    return href
#获取所有页的url列表
result = dict()
pages = get_page(target, pagenum)
for html in pages:
    url = get_url(html)
    result.update(url)
#通过url查找到公司的地址信息
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.87 Safari/537.36'}
import time
for i in range(len(result)):
    url=list(result.values())[i]
    key=list(result.keys())[i]
    print (key,url)
    r=requests.get(url, headers=headers)
    html = r.text
    #print (html)
    bf = BeautifulSoup(html)
    addr = bf.find('p', class_='mlist_li_desc')
    #print(addr)
    print (addr.get_text().strip())
    addr_list[key] = addr.get_text().strip()
    time.sleep(2)
#保存结果到txt文件
file = open('/Users/caoshishi/Downloads/addr_list.txt', 'w')
for item in addr_list.items():
    item = ','.join(list(item))
    file.write(item)
    file.write('\n')
file.close()
```
